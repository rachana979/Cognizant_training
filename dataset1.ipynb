```python
empolyee={
    "name":["stya","soumii","ramesh","shivv","visruu"],
    "salary":[100000,90000,80000,70000,60000]}
result=pd.DataFrame(empolyee)
print(result[result["salary"]>80000])
```

```python
empolyee={
    "name":["stya","soumii","ramesh","shivv","visruu"],
    "salary":[100000,90000,80000,70000,60000]}
result=pd.DataFrame(empolyee)
print(result.query("salary>80000"))
```

```python
empolyee={
    "name":["stya","soumii","ramesh","shivv","visruu"],
    "salary":[100000,90000,80000,70000,60000]}
result=pd.DataFrame(empolyee)
print(result.query("salary>80000 and salary<95000"))
```
