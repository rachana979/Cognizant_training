```python
import matplotlib.pyplot as plt
import numpy as np
products=["samsung","iphone","andorid"]
year_2020=[2000,300,400]
year_2021=[5000,600,702]
year_2022=[8000,900,101]
x=np.arange(len(products))
width=0.35
plt.bar(x-width/2,year_2020,width,label='2020')
plt.bar(x-width/2,year_2021,width,label='2021')
plt.bar(x-width/2,year_2022,width,label='2022')
plt.xlabel("products")
plt.title("year comparision: 2020 vs 2021 ")
plt.xticks(x,products)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
products=["samsung","iphone","andorid"]
online=[2000,300,4000]
offline=[5000,600,7000]
wholesale=[8000,900,1010]
plt.bar(products,online,label='online')
plt.bar(products,offline,bottom=online,label='offline')
bottom_value=[online[i]+offline[i] for i in range(len(products))]
plt.bar(products,wholesale,bottom=bottom_value,label='wholesale')
plt.xlabel("products")
plt.ylabel("Revenue")
plt.title("Revenue of the Distrubted sales")
plt.legend()
plt.xticks(x,products)
plt.show()
```
