```python
import seaborn as sns
import matplotlib.pyplot as plt

months=["jan","feb","mar","apr","may"]
sales=[200,199,300,250,400]
sns.lineplot(x=months,y=sales)
plt.xlabel("months")
plt.ylabel("sales")
plt.title("monthly sales trends")
plt.show()
```

```python
import plotly.express as px
import pandas as pd
df=pd.DataFrame({
    'semster':['sem_1','sem_2','sem_3','sem_4'],
    'marks':[100,95,120,110]})
fig=px.line(df,x='semster',y='marks',markers=True)
fig.show()
```
