```python
clusters={
    "Hyd":[1,2,3,4,5],
    "Pune":[6,7,8,9,10],
    "Chennai":[11,122,13,14,15]
}
select_clusters=["Hyd","Chennai"]
sample=[] 
for cluster in select_clusters:
    sample.extend(clusters[cluster])
    print("cluster sample:",sample)
```

```python
customer=list(range(1,100))
sample=customer[::10]
print("Convenience Sample",sample)
```

```python
empolyee={
    "A":10,
    "B":20,
    "C":30,
    "D":40
}
sample=[name for name,exp in empolyee.items() if exp>=10]
print("Judgment sample:",sample)
```

```python
network={
    "A":["B","C",],
    "B":["D"],
    "C":["E"]
}
sample=["A"]
sample.extend(network["A"])
sample.extend(network["B"])
print("snowBall sampling:",sample)
```
